```python
import numpy as np
from matplotlib import pyplot as plt

b = 2               # 周期的个数
m = 100             # training set 中元素的个数
l = 100
x0 = np.linspace(0, b, l)
y0 = np.sin(2 * np.pi * x0)
x = np.linspace(0, b, m)

y = np.zeros(m)
for i in range(m):
    y[i] = np.sin(2 * np.pi * x[i])
    y[i] =  y[i] + np.random.normal(0, .8)  # 加入 Gauss 分布的噪声
plt.plot(x0, y0, 'gray')
plt.scatter(x, y)

yf = np.zeros(l)
n = 10              # 多项式拟合的最高阶
a = np.polyfit(x, y, n)
for i in range(l):
    for j in range(n+1):
        yf[i] = yf[i] + a[n-j]*x0[i]**j
plt.plot(x0, yf, 'r')
```
